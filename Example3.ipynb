# Example 3: Binary Mixture with a Tangent Pinch

#### This example shows the distillation column for a mixture of Benzene and Ethylenediamine

Like Example1, this is a binary mixture, with the Reflux Ratio as the primary factor determining column feasability,  However, the shape of the equilibrium cuvre leads to an additional constraint on the Reflux Ratio due at a pinch point -- a new type of Fixed Point

```python
## IMPORT NECESSARY LIBRARIES ##
import numpy as np
import matplotlib.pyplot as plt

from src.distillation.DistillationBinary import DistillationModelBinary

import ipywidgets as widgets
from IPython.display import display
from src.thermo_models.VLEEmpiricalModel import *
from ipywidgets import interactive, Button, VBox, HBox, Output
from IPython.display import clear_output

import seaborn as sns

plt.rcParams.update({"text.usetex": True,"font.family": "sans-serif","font.sans-serif": ["Helvetica"]})
```

### Demo 1: Equilibrium Curve Shape

**Note the difference in the shape of this Equilibrium Curve.  In Example 1, the curve was Concave-Down over the whole range.  However, in this example, the curve has an inflection point.  This inflection can cause a tangent pinch between the Equlibrium Curve and Rectifying Line, which will be explored in this Example.**

```python
#Define empirical relationship 
def xtoy(x, a = 9, b = -0.6):
    y = a*x/(1+(a-1)*x) +b*x*(1 - x)
    return y
    
# Create the VLE model object
vle_model = VLEEmpiricalModelBinary(["Benzene", "Ethylenediamene"], xtoy)
    
vle_model.plot_binary_yx()    
```

```python
def plot_distil_oplines_ex2(dist_model, xD, xB, R):
    
    dist_model.set_xD(np.array([xD, 1 - xD]))
    dist_model.set_xB(np.array([xB, 1 - xB]))
    dist_model.set_r(R)

    fig, axs = plt.subplots(1, 3, figsize=(14, 6))
    
    dist_model.plot_distil_strip_binary(axs[0])
    dist_model.plot_distil_rect_binary(axs[1])
    dist_model.plot_distil_binary(axs[2])
    
    plt.tight_layout()
    plt.show()  # Ensure the plot is displayed in Jupyter.

def plot_distil_stripping_ex2(dist_model, xD, xB, R):
    
    dist_model.set_xD(np.array([xD, 1 - xD]))
    dist_model.set_xB(np.array([xB, 1 - xB]))
    dist_model.set_r(R)
    
    fig, axs = plt.subplots(1, 1, figsize=(7, 7))
    
    dist_model.plot_distil_strip_binary(ax=axs)
    
    plt.tight_layout()
    plt.show()  # This is essential for the plot to be displayed in the Output widget in Jupyter.

def plot_distil_rectifying_ex2_demo2(dist_model, xD, xB, R):
    
    dist_model.set_xD(np.array([xD, 1 - xD]))
    dist_model.set_xB(np.array([xB, 1 - xB]))
    dist_model.set_r(R)
    
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    dist_model.plot_distil_rect_binary(ax=axs[0])
    dist_model.plot_distil_rect_binary(ax=axs[1], rect_title = "Zoomed in View of Tangent Pinch")
    
    axs[1].set_ylim([0.8, 1.0])
    axs[1].set_xlim([0.8, 1.0])
    
    plt.tight_layout()
    plt.show() 

    
def plot_distil_rectifying_ex2_demo3(dist_model, xD, xB, R):
    
    dist_model.set_xD(np.array([xD, 1 - xD]))
    dist_model.set_xB(np.array([xB, 1 - xB]))
    dist_model.set_r(R)
    
    fig, axs = plt.subplots(1, 1, figsize=(7, 7))
    
    dist_model.plot_distil_rect_binary(ax=axs)
    
    plt.tight_layout()
    plt.show()  
```

```python

# Creating a distillation model object
distillation_model = DistillationModelBinary(vle_model, xF=np.array([0.5, 0.5]), xD=np.array([0.999, 0.001]), xB=np.array([0.001, 0.999]), reflux=2)
```

```python
# Define the sliders:
xD1_slider = widgets.FloatSlider(value=0.95, min=0.5,   max=0.999, step=0.001, description='$x_D$:',        readout_format='.3f')
xB1_slider = widgets.FloatSlider(value=0.05, min=0,     max=0.5,   step=0.001, description='$x_B$:',        readout_format='.3f')
R1_slider  = widgets.FloatSlider(value=2.75, min =1,    max=3.5,   step=0.01,  description='Reflux Ratio:', readout_format='.2f')
R2_slider  = widgets.FloatSlider(value=2.75, min=1,     max=3.5,   step=0.01,  description='Reflux Ratio:', readout_format='.2f')


# Setup the interactive objects
out_strip   = Output()
out_rect2   = Output()
out_rect3   = Output()
out_oplines = Output()
```

###  Demo 2: Rectifying Pinch Region

The inflection point in the equilibrum curve restricts the range of feasible reflux rations when the Distillate Composition $x_D$ is sufficiently high (set at 0.99 for this Demo).  This demo shows the full graph, and then a zoomed in view of the tangent pinch region

Move the Reflux Ratio to see how it affects the number of stages near the pinch point.  Adjust the reflux ratio to values below 1.7. At these reflux ratios the operating line is almost tangential or tangential to the y-x curve and therefore the number of stages dramatically increases. The compositions of the liquid that exits the first 200 stages are shown as blue triangles along the x-axis

*The figure to the right is a zoomed in version of the left graph near the inflection point*

```python
# Define your button click behavior
def on_rect_button_click(b):
    with out_rect2:
        clear_output()
        plt.close('all')  # Close all open matplotlib plots
        plot_distil_rectifying_ex2_demo2(distillation_model, xD=0.99, xB=0.001, R=R2_slider.value)

rect_button = Button(description='Generate Plot')
rect_button.on_click(on_rect_button_click)

layout = VBox([R2_slider, rect_button, out_rect2])
display(layout)
```

### Demo 3: Low Distillate Purity

While the Reflux Ratio may seem like the only parameter that determines feasability, the Distillate Composition $x_D$ can also determine feasability.

Here $x_D$ is set to equal 0.95, making the tangent pinch no longer a concern. Changing the column parameters to move away from the tangent pinch region is a valid method for avoiding column infeasability.

Note that the minimum reflex ratio in this case is much lower than the previous case

```python
# Define your button click behavior
def on_rect_button_click(b):
    with out_rect3:
        clear_output(wait=True)
        plt.close('all')  # Close all open matplotlib plots
        plot_distil_rectifying_ex2_demo3(distillation_model, xD=0.95, xB=0.05, R=R1_slider.value)

rect_button = Button(description='Generate Plot')
rect_button.on_click(on_rect_button_click)

layout = VBox([R1_slider, rect_button, out_rect3])
display(layout)
```

### Demo 4: Full Distillation Column

This demonstration shows the previous graphs, as well as a third graph on the right for the entire distillation column.  The color of the operating lines shows changes based on the set parameters.  Green operating lines indicate a feasible column.  Black operating lines mean the column is not feasible due to a tangent pinch.  Try to adjust the parameters to create such a column.  

How many stages do you need if the distillate composition is set to 0.99 and reflux ratio to 1.65 ? What happens as you reduce distillate composition to 0.95

**Red operating lines also mean the column is not feasible due to a reflux ratio below the minimum possible value.  However, these were not caused by the tangent pinch point of the graph.  How many fixed points do you count for these columns?**

**Note that not all of the symbols on the x-axis correspond to compositions of liquid phase that exits a stage in the column.**

```python
# Define your button click behavior
def on_oplines_button_click(b):
    with out_oplines:
        clear_output()
        plt.close('all')  # Close all open matplotlib plots
        plot_distil_oplines_ex2(distillation_model, xD=xD1_slider.value, xB=xB1_slider.value, R=R1_slider.value)

oplines_button = Button(description='Generate Plots')
oplines_button.on_click(on_oplines_button_click)

layout_oplines = VBox([xD1_slider, xB1_slider, R1_slider, oplines_button, out_oplines])
display(layout_oplines)
```
